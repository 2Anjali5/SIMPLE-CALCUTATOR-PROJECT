# SIMPLE CALCULATOR

```python
import tkinter
from tkinter import *
window=Tk()
window.title(" MY CALCULATOR")
window.geometry("370x448")
window.configure(bg='black')
window.resizable(False,False)

equation=" "

def show(value):
    global equation
    equation+=value
    label.config(text=equation)

def clear():
    global equation
    equation=""
    label.config(text=equation)
    
def calculate():
    global equation
    result=""
    if equation!="":
        try:
            result=eval(equation)
        except:
            result="error"
            equation=""    
    label.config(text=result) 
    
label=Label(window,width=25,height=2,bg="grey",text=" ",bd=1,fg='#FA03D1',font=("arial",20))
label.pack()

button1=Button(window,width=9,height=3,text="C",bg='#03E4FA',command=lambda:clear())
button1.place(x=10,y=100)

button2=Button(window,width=9,height=3,text="/",command=lambda:show("/"))
button2.place(x=100,y=100)

button3=Button(window,width=9,height=3,text="%",command=lambda:show("%"))
button3.place(x=190,y=100)

button4=Button(window,width=9,height=3,text="*",command=lambda:show("*"))
button4.place(x=280,y=100)

button5=Button(window,width=9,height=3,text="7",command=lambda:show("7"))
button5.place(x=10,y=170)

button6=Button(window,width=9,height=3,text="8",command=lambda:show("8"))
button6.place(x=100,y=170)

button7=Button(window,width=9,height=3,text="9",command=lambda:show("9"))
button7.place(x=190,y=170)

button8=Button(window,width=9,height=3,text="-",command=lambda:show("-"))
button8.place(x=280,y=170)

button9=Button(window,width=9,height=3,text="4",command=lambda:show("4"))
button9.place(x=10,y=240)

button10=Button(window,width=9,height=3,text="5",command=lambda:show("5"))
button10.place(x=100,y=240)

button11=Button(window,width=9,height=3,text="6",command=lambda:show("6"))
button11.place(x=190,y=240)

button12=Button(window,width=9,height=3,text="+",command=lambda:show("+"))
button12.place(x=280,y=240)

button13=Button(window,width=9,height=3,text="1",command=lambda:show("1"))
button13.place(x=10,y=310)

button14=Button(window,width=9,height=3,text="2",command=lambda:show("2"))
button14.place(x=100,y=310)

button15=Button(window,width=9,height=3,text="3",command=lambda:show("3"))
button15.place(x=190,y=310)

button16=Button(window,width=9,height=7,text="=",bg="#EC9D00",command=lambda:calculate())
button16.place(x=280,y=310)

button17=Button(window,width=22,height=2,text="0",command=lambda:show("0"))
button17.place(x=10,y=380)

button18=Button(window,width=9,height=2,text=".",command=lambda:show("."))
button18.place(x=190,y=380)

window.mainloop()
```
